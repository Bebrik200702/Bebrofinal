```python
from tqdm.auto import tqdm
import supervision as sv
import pycocotools
import json
import pycocotools.mask
import numpy as np
from ultralytics import YOLO
import albumentations as A
import os
model = YOLO('runs/segment/train23/weights/best.pt')

test_folder = "test"
output_folder = "output_yolo"  # –ü–∞–ø–∫–∞ –¥–ª—è —Å–æ—Ö—Ä–∞–Ω–µ–Ω–∏—è —Ä–µ–∑—É–ª—å—Ç–∞—Ç–æ–≤
os.makedirs(output_folder, exist_ok=True)

# –°–ø–∏—Å–æ–∫ –¥–ª—è —Ö—Ä–∞–Ω–µ–Ω–∏—è –ø—Ä–µ–¥—Å–∫–∞–∑–∞–Ω–∏–π –≤ –Ω—É–∂–Ω–æ–º —Ñ–æ—Ä–º–∞—Ç–µ
predictions = []

# –û–±—Ä–∞–±–æ—Ç–∫–∞ –∏–∑–æ–±—Ä–∞–∂–µ–Ω–∏–π
# –°–ø–∏—Å–æ–∫ –¥–ª—è —Ö—Ä–∞–Ω–µ–Ω–∏—è –ø—Ä–µ–¥—Å–∫–∞–∑–∞–Ω–∏–π –≤ –Ω—É–∂–Ω–æ–º —Ñ–æ—Ä–º–∞—Ç–µ
predictions = []
# –û–±—Ä–∞–±–æ—Ç–∫–∞ –∏–∑–æ–±—Ä–∞–∂–µ–Ω–∏–π
for file_name in tqdm(os.listdir(test_folder)):
    image_path = os.path.join(test_folder, file_name)
    cnf = 0.5
    results = model(image_path,conf=cnf,verbose=False)[0]
    if results.masks is None:
        continue
    outputs = sv.Detections.from_ultralytics(results)#.with_nms(threshold=0.5, class_agnostic=False)
    pred_classes = outputs.class_id.tolist()  # –ö–ª–∞—Å—Å—ã –æ–±—ä–µ–∫—Ç–æ–≤
    pred_boxes = outputs.xyxy.tolist()#instances.pred_boxes.tensor.tolist()  # –ö–æ–æ—Ä–¥–∏–Ω–∞—Ç—ã bbox
    pred_scores = outputs.confidence.tolist()  # –°–∫–æ—Ä–∏–Ω–≥ –ø—Ä–µ–¥—Å–∫–∞–∑–∞–Ω–∏–π
    pred_segmentation = outputs.mask.tolist()#instances.pred_masks.tolist()  # –°–µ–≥–º–µ–Ω—Ç–∞—Ü–∏—è
        
    # –°–æ—Ö—Ä–∞–Ω—è–µ–º –ø—Ä–µ–¥—Å–∫–∞–∑–∞–Ω–∏—è –¥–ª—è –∫–∞–∂–¥–æ–≥–æ –æ–±—ä–µ–∫—Ç–∞ –≤ –∏–∑–æ–±—Ä–∞–∂–µ–Ω–∏–∏
    for i in range(len(pred_classes)):
        # pred_segmentation = instances.pred_masks.tolist()
        rle = pycocotools.mask.encode(np.asfortranarray(pred_segmentation[i]))
        predictions.append({
            "image_name": file_name,
            "category_id": int(pred_classes[i]),  # ID –∫–∞—Ç–µ–≥–æ—Ä–∏–∏
            "bbox": pred_boxes[i],  # –ö–æ–æ—Ä–¥–∏–Ω–∞—Ç—ã bounding box
            "score": pred_scores[i],  # –û—Ü–µ–Ω–∫–∞ —É–≤–µ—Ä–µ–Ω–Ω–æ—Å—Ç–∏
            "segmentation": {
                "size" : rle["size"],
                "counts" : str(rle["counts"])
            }  # –°–µ–≥–º–µ–Ω—Ç–∞—Ü–∏—è –≤ —Ñ–æ—Ä–º–∞—Ç–µ RLE
        })


# –°–æ—Ö—Ä–∞–Ω–µ–Ω–∏–µ –ø—Ä–µ–¥—Å–∫–∞–∑–∞–Ω–∏–π –≤ JSON
predictions_file = os.path.join(output_folder, "submission.json")
with open(predictions_file, "w") as f:
    json.dump(predictions, f, indent=4)

print(f"–û–±—Ä–∞–±–æ—Ç–∫–∞ –∑–∞–≤–µ—Ä—à–µ–Ω–∞. –†–µ–∑—É–ª—å—Ç–∞—Ç—ã —Å–æ—Ö—Ä–∞–Ω–µ–Ω—ã –≤ –ø–∞–ø–∫–µ: {output_folder}")
```

```python
# –°–æ—Ö—Ä–∞–Ω–µ–Ω–∏–µ –ø—Ä–µ–¥—Å–∫–∞–∑–∞–Ω–∏–π –≤ JSON
predictions_file = os.path.join(output_folder, "submission.json")
with open(predictions_file, "w") as f:
    json.dump(predictions, f, indent=4)

print(f"–û–±—Ä–∞–±–æ—Ç–∫–∞ –∑–∞–≤–µ—Ä—à–µ–Ω–∞. –†–µ–∑—É–ª—å—Ç–∞—Ç—ã —Å–æ—Ö—Ä–∞–Ω–µ–Ω—ã –≤ –ø–∞–ø–∫–µ: {output_folder}")
```
